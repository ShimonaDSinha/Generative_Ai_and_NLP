```python
import os
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['USE_TF'] = 'NO'
from transformers import pipeline
```

```python
qa_pipeline = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)
```

```python
qa_pipeline = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)
```

```python
context = """
Generative AI poses several risks including hallucinations,
algorithmic bias, misinformation, data privacy concerns,
and the creation of deepfake content.
"""
```

```python
question = "What are the risks of Generative AI?"
```

```python
result = qa_pipeline(
    question=question,
    context=context
)

print("Answer:", result["answer"])
print("Confidence Score:", result["score"])
```

```python
questions = [
    "What are the risks?",
    "Does Generative AI cause bias?",
    "What content can be misused?"
]

for q in questions:
    output = qa_pipeline(question=q, context=context)
    print("\nQuestion:", q)
    print("Answer:", output["answer"])
```

**Architecture Explanation**

Why this model works?

DistilBERT is an encoder-only transformer architecture trained to understand context and predict answer spans within a document.
Because it is fine-tuned on the SQuAD dataset, it performs extractive Question Answering effectively by identifying the most relevant text segment.

**Limitation**
This model cannot generate new answers. It can only extract answers that already exist in the given context. If the answer is not present, the output may be incomplete or incorrect.
